### Analyse De Corr√©lation :


```python

# Importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the file
df = pd.read_csv('../spotify.csv', encoding='latin1')
# print(df.head())
print(df)
```

```python
# Convertir la colonne 'streams' en un type de donn√©es num√©rique
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# V√©rifier s'il y a des donn√©es manquantes dans la colonne 'streams'
print(df['streams'].isnull().sum())

# Calculer la popularit√© moyenne des chansons par ann√©e de sortie
popularity_by_year = df.groupby('released_year')['streams'].mean()

# Cr√©ation du graphique √† barres
plt.figure(figsize=(10, 6))
popularity_by_year.plot(kind='bar', color='skyblue')

# Ajout de titres et de labels
plt.title('Popularit√© des chansons par ann√©e de sortie')
plt.xlabel('Ann√©e de sortie')
plt.ylabel('Popularit√© moyenne (Streams)')
plt.xticks(rotation=45)  # Rotation des √©tiquettes de l'axe x pour une meilleure lisibilit√©
plt.grid(axis='y')

# Affichage du graphique
plt.tight_layout()
plt.show()
```
